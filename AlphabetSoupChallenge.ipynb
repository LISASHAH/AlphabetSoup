```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,OneHotEncoder
import pandas as pd
import tensorflow as tf
```

```python
charityDataDF=pd.read_csv('charity_data.csv')
```

```python
charityDataDF.head()
```

```python
charityPreProcDF=charityDataDF.copy()
# Filters out non-feature variables from dataset
charityPreProcDF=charityPreProcDF.drop(columns=['NAME','EIN'])
charityPreProcDF.head()
```

```python
charityCat = charityPreProcDF.dtypes[charityPreProcDF.dtypes == "object"].index.tolist()
```

```python
charityPreProcDF[charityCat].nunique()
```

```python
#Applies categorical binning for APPLICTION_TYPE
applicationCounts=charityPreProcDF.APPLICATION_TYPE.value_counts()
replace_type=list(applicationCounts[applicationCounts<500].index)
for application in replace_type:
    charityPreProcDF.APPLICATION_TYPE = charityPreProcDF.APPLICATION_TYPE.replace(application,"Others")
charityPreProcDF.head()   
```

```python
charityPreProcDF.APPLICATION_TYPE.value_counts()
```

```python
#Applies categorical binning for CLASSIFICATION
classificationCounts=charityPreProcDF.CLASSIFICATION.value_counts()
replace_classification=list(classificationCounts[classificationCounts<1000].index)
for classification in replace_classification:
    charityPreProcDF.CLASSIFICATION = charityPreProcDF.CLASSIFICATION.replace(classification,"Others")
charityPreProcDF.head() 
```

```python
charityPreProcDF.CLASSIFICATION.value_counts()
```

```python
# Create a OneHotEncoder instance
enc = OneHotEncoder(sparse=False)
# Fit and transform the OneHotEncoder using the categorical variable list
encode_df = pd.DataFrame(enc.fit_transform(charityPreProcDF[charityCat]))
# Add the encoded variable names to the dataframe
encode_df.columns = enc.get_feature_names(charityCat)
encode_df.head()
```

```python
charityPreProcDF = charityPreProcDF.merge(encode_df,left_index=True, right_index=True)
```

```python
(charityPreProcDF.columns)
```

```python
y=charityPreProcDF.IS_SUCCESSFUL.values
```

```python
#Lisa-are these the correct X columns
X=charityPreProcDF.drop(columns=['APPLICATION_TYPE',
 'AFFILIATION',
 'USE_CASE',
 'CLASSIFICATION',
 'ORGANIZATION',
 'INCOME_AMT',
 'SPECIAL_CONSIDERATIONS','IS_SUCCESSFUL'])
```

```python
len(X.columns)
```

```python
# Split training/test datasets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
# Create a StandardScaler instance
scaler = StandardScaler()
# Fit the StandardScaler
X_scaler = scaler.fit(X_train)
# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
#Define a neural network model using Tensorflow Keras module
#Define the number of layers and number of neurons per layer
number_input_features = len(X_train_scaled[0])
hidden_nodes_layer1 =  100
hidden_nodes_layer2 = 10
#Defines a neural network model
nn = tf.keras.models.Sequential()
```

```python
# First hidden layer
nn.add(
    tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation="relu")
)
# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation="relu"))
# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))
# Compile the Sequential model together and customize metrics
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model 
fit_model = nn.fit(X_train_scaled, y_train, epochs=100) 
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
